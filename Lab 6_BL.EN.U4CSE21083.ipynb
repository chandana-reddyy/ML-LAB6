```python
#A1
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('the_final_no_changes_data.csv')

x_values = data['len']
y_values = data['subjectivity']

plt.figure(figsize=(8,6))
plt.scatter(x_values, y_values, color='r', alpha=0.7)
plt.title('Scatter Plot of Length vs Subjectivity')
plt.xlabel('Length')
plt.ylabel('Subjectivity')
plt.grid(True)
plt.show()
```

```python
#A2
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data = pd.read_csv('the_final_no_changes_data.csv')

X = data['polarity'].values.reshape(-1, 1)
Y = data['subjectivity'].values

model = LinearRegression()
model.fit(X, Y)

predicted_Y = model.predict(X)

mse = mean_squared_error(Y, predicted_Y)
print(f"Mean Squared Error: {mse:.2f}")
```
