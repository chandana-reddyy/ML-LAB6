```python
#A2
# Importing necessary libraries 
import pandas as pd
# Importing necessary libraries for linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
#loading dataset
data = pd.read_csv('the_final_no_changes_data.csv')
# Extracting the subjectivity feature as X and reshaping it to a 2D array
X = data['subjectivity'].values.reshape(-1, 1)
# Extracting the text feature as y
y = data['text'].values
# Splitting the dataset into training and testing sets wth trainseize 0.8 and test size 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# Creating a Linear Regression model
model = LinearRegression()
# Fitting the model on the training data
model.fit(X_train, y_train)
# Making predictions on the test data
y_pred = model.predict(X_test)
# Calculating mse
mse = mean_squared_error(y_test, y_pred)
# Print the  value of mse
print(f"Mean Squared Error: {mse}")
```

```python
#A3
# Extract independent and dependent features
independent_feature = data['text'].values.reshape(-1, 1)
dependent_feature = data['category'].values
# taking model coefficients and attributes
regression_coefficients = model.coef_
regression_intercept = model.intercept_
matrix_rank = model.rank_
singular_vals = model.singular_
num_selected_features = model.n_features_in_
# Making predictions for a new data point
new_feature_value = [[10]]
predicted_value = model.predict(new_feature_value)
# Calculating  Mean Squared Error 
mse = mean_squared_error(dependent_feature, model.predict(independent_feature))
# Calculating R-squared 
r_squared = model.score(independent_feature, dependent_feature)
# printing the results the results
print("Regression Coefficients:", regression_coefficients)
print("Regression Intercept:", regression_intercept)
print("Rank of Feature Matrix:", matrix_rank)
print("Singular Values of Feature Matrix:", singular_vals)
print("Number of Features Used:", num_selected_features)
print("Predicted Value for New Feature:", predicted_value)
print("Mean Squared Error (MSE):", mse)
print("R-squared (Model Score):", r_squared)
```

```python
#A4
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Loading dataset
data = pd.read_csv('the_final_no_changes_data.csv')
# Defining a threshold for binary classification
threshold = 0.5
# Creating binary labels based on threshold value
data['binary_class'] = (data['polarity'] >= threshold).astype(int)
# Selecting features and binary labels
X = data[['subjectivity', 'len']]
y = data['binary_class']
# Splitting the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# Creating a Logistic Regression classifier
logistic_classifier = LogisticRegression()
# Training the classifier on the training set
logistic_classifier.fit(X_train, y_train)
# Making predictions on the test set
y_pred = logistic_classifier.predict(X_test)
# Calculating the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
# Printing the accuracy
print(f"Accuracy on the Test Set: {accuracy}")
```

```python
#A5
#importing libraries required for decision tree  and knn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
# Loading dataset
data = pd.read_csv('the_final_no_changes_data.csv')
# Selecting features and target variable
features = data[['polarity', 'subjectivity']]
target = data['category']
# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
# Training a Decision Tree Regressor
decision_tree_regressor = DecisionTreeRegressor()
decision_tree_regressor.fit(X_train, y_train)
decision_tree_predictions = decision_tree_regressor.predict(X_test)
decision_tree_mse = mean_squared_error(y_test, decision_tree_predictions)
print(f"Mean Squared Error for Decision Tree Regressor: {decision_tree_mse:.2f}")
# Training a k-Nearest Neighbors Regressor
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(X_train, y_train)
knn_predictions = knn_regressor.predict(X_test)
knn_mse = mean_squared_error(y_test, knn_predictions)
print(f"Mean Squared Error for k-NN Regressor: {knn_mse:.2f}")
```
